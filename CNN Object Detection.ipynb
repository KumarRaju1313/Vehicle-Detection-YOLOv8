```python
import os
import shutil
from sklearn.model_selection import train_test_split

image_folder = '/content/drive/MyDrive/Colab Notebooks/Car_Class (1)/Frames'
label_folder = '/content/drive/MyDrive/Colab Notebooks/Car_Class (1)/Labels'
output_image_train = '/content/drive/MyDrive/dataset/images/train'
output_image_val = '/content/drive/MyDrive/dataset/images/val'
output_label_train = '/content/drive/MyDrive/dataset/labels/train'
output_label_val = '/content/drive/MyDrive/dataset/labels/val'

os.makedirs(output_image_train, exist_ok=True)
os.makedirs(output_image_val, exist_ok=True)
os.makedirs(output_label_train, exist_ok=True)
os.makedirs(output_label_val, exist_ok=True)

all_images = [f for f in os.listdir(image_folder) if f.endswith('.jpg') or f.endswith('.png')]

train_images, val_images = train_test_split(all_images, test_size=0.2, random_state=42)

for img in train_images:
    shutil.move(os.path.join(image_folder, img), os.path.join(output_image_train, img))
    label_file = img.replace('.jpg', '.txt').replace('.png', '.txt')
    shutil.move(os.path.join(label_folder, label_file), os.path.join(output_label_train, label_file))

for img in val_images:
    shutil.move(os.path.join(image_folder, img), os.path.join(output_image_val, img))
    label_file = img.replace('.jpg', '.txt').replace('.png', '.txt')
    shutil.move(os.path.join(label_folder, label_file), os.path.join(output_label_val, label_file))
```

```python
import os
```

```python
label_folder = '/content/drive/MyDrive/dataset/labels/train'
val_label_folder = '/content/drive/MyDrive/dataset/labels/val'

original_class_id = 2
new_class_id = 0

for label_file in os.listdir(label_folder):
    if label_file.endswith('.txt'):
        with open(os.path.join(label_folder, label_file), 'r') as f:
            lines = f.readlines()

        new_lines = []
        for line in lines:
            parts = line.split()
            if parts:
                if int(parts[0]) == original_class_id:
                    parts[0] = str(new_class_id)
                new_lines.append(' '.join(parts) + '\n')

        with open(os.path.join(label_folder, label_file), 'w') as f:
            f.writelines(new_lines)

for label_file in os.listdir(val_label_folder):
    if label_file.endswith('.txt'):
        with open(os.path.join(val_label_folder, label_file), 'r') as f:
            lines = f.readlines()

        new_lines = []
        for line in lines:
            parts = line.split()
            if parts:
                if int(parts[0]) == original_class_id:
                    parts[0] = str(new_class_id)
                new_lines.append(' '.join(parts) + '\n')

        with open(os.path.join(val_label_folder, label_file), 'w') as f:
            f.writelines(new_lines)
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('/content/runs/detect/train5/weights/best.pt')

input_video_path = '/content/output_video4.mp4'
output_video_path = '/content/output_video1.mp4'

cap = cv2.VideoCapture(input_video_path)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while True:
    ret, frame = cap.read()
    if not ret:
        break
    results = model.predict(frame)
    annotated_frame = results[0].plot()
    out.write(annotated_frame)
cap.release()
out.release()
cv2.destroyAllWindows()
```
